In this part, you will be working with the library ```pandas```. Import this library with an alias name ```pd```.

````
# Initial imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
````

Download dataset from: http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv
Read the data and display the first 5 rows

````
#Download dataset from: http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv

file_url = "http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv"
data = pd.read_csv(file_url)
data.head()
````

### Preprocessing of the Dataset



---


**1. Convert the categorical values of columns 'sex', 'embarked', 'pclass' into dummy values and add them into titanic_data_new DataFrame**




---



````
data["sex"] = pd.get_dummies(data["sex"])
data["embarked"] = pd.get_dummies(data["embarked"])
data["pclass"] = pd.get_dummies(data["pclass"])
titanic_data_new = pd.DataFrame(data, columns=["sex","embarked","pclass"])
````



---


**2. Fill the missing data in Age and Fare columns with the mean value of the respected column then add these columns to titanic_data_new**


---



````
meanAge = np.mean(data["age"],axis = 0)
meanFare = np.mean(data["fare"],axis = 0)
data["age"].fillna(meanAge, inplace = True)
data["fare"].fillna(meanFare, inplace=True)
titanic_data_new["age"] = data["age"]
titanic_data_new["fare"] = data["fare"]
````

###Normalization


---



1. Normalize the columns which have numerical data



---



````
normalized = []
for col in data.columns:    
  try:
    data[col] = (data[col]-min(data[col]))/(max(data[col])-min(data[col]))
    normalized.append(col)
  except TypeError:
    continue
  
````



---
2. Drop the original age, fare columns
3. Append the normalized columns to the titanic_data_new DataFrame.



---



````
dataB = pd.DataFrame(data)
data.drop(columns = ["age","fare"],axis=0,inplace=True)
for col in normalized:
  if(col!="age" and col!="fare"):
    titanic_data_new[col] = data[col]
````

###Plotting
1. plot the age against the ID(continuous)
2. plot the fare against the ID(scatter)

````
f = plt.figure(figsize=(10,5))
ax1 = f.add_subplot(121)
ax2 = f.add_subplot(122)
ax1.plot(dataB["age"])
id = np.arange(len(dataB["fare"]))
ax2.scatter(id,dataB["fare"])
````

Plot the sine function with domain in $[-3\pi,3\pi]$. Label the axes appropriately. Add title $\sin(x)$. Use latex formatting while adding the title. Set the plot size to (10,10)

````
x = np.arange(-np.pi*3,np.pi*3,np.pi/20)
y = np.sin(x)
f = plt.figure(figsize=(10,10))
plt.plot(x,y)
plt.title(r'$\sin(x)$')
plt.show()
````

Plot $\sin(x)$ and its spectrum, $\cos(x)$ and its spectrum in a single plot, in the domain $[-2\pi,2\pi]$. Label the plot appropriately.

````
x = np.arange(-np.pi*3,np.pi*3,np.pi/20)
y = np.sin(x)
f = plt.figure(figsize=(10,5))
ax1 = f.add_subplot(121)
ax1.plot(x,y)
plt.title('sin(x)')
plt.xlabel('Time')
plt.ylabel('Amplitude')
spec = np.fft.fft(y)
ax2 = f.add_subplot(122)
ax2.plot(abs(spec))
plt.title('Spectrum of sin(x)')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
````
